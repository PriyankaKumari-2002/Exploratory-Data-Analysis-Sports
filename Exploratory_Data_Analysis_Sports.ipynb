#To Find the teams who chose batting or balling in Feroz Shah Kotla Stadium won maxium number of times?

```python
#stadium wise analysis: checking whether batting or bowling first has more advantage
delhi=match_df.loc[(match_df['venue']=='Feroz Shah Kotla')]
# teams win by runs must have chosen batting first
delhirun=delhi[delhi['win_by_runs']>0]
chart=[len(delhirun),len(delhi)-len(delhirun)]
labels=['Batting First','Balling First']
plt.pie(chart,labels=labels,startangle=20,explode=(0,0.2),shadow=2,colors=['red','purple'])
plt.show()
```

**Observation** In Feroz Shah Kotla Stadium located in delhi batting or balling first doesnt mater as both have equal probability of wining the match

#To Analyse how toss affected win or loss of a team?

```python
import numpy as np
toss = match_df.groupby([ 'toss_winner']).winner.value_counts().reset_index(name = 'count')
toss['result'] = np.where(toss.toss_winner == toss.winner, 'won', 'lost')
toss_result = toss.groupby([ 'toss_winner','result'])['count'].sum().reset_index()

plot = sns.barplot(x="toss_winner", y="count", hue="result", data=toss_result)
plot.set_title('Matches won/lost by teams winning toss \nSeason ')
```

```python
batsman_grp = del_df.groupby(["match_id", "inning", "batting_team", "batsman"])
batsmen = batsman_grp["batsman_runs"].sum().reset_index()
batsmen = match_df[['id','season']].merge(batsmen, left_on = 'id', right_on = 'match_id', how = 'left').drop('id', axis = 1)
```

```python
batsmen
```

#Finding Top Scoring Batsmen and the scores they have made

```python
batsman_runsperseason = batsmen.groupby(['season', 'batting_team', 'batsman'])['batsman_runs'].sum().reset_index() #this groupby to have a single value
#per batsmen for a season
batsman_runsperseason = batsman_runsperseason.groupby(['season', 'batsman'])['batsman_runs'].sum().unstack().T
batsman_runsperseason['Total'] = batsman_runsperseason.sum(axis=1)
batsman_runsperseason = batsman_runsperseason.sort_values(by = 'Total', ascending = False).drop('Total', 1)
ax = batsman_runsperseason[:7].T.plot()
```

**Observation** 1. DA warner was improving every forecoming ipl and performed his best in 2016 but didnt play in 2018 ipl 
2. Sk Raina has been a consistent and showed constant performance in every ipl 
3. GH gayle scored best in ipl 2012

```python
#dictionary to get winners of each season
winning_teams=match_df[['season','winner']] #creating df with season and winner as columns
winners_team={}
for i in sorted(winning_teams.season.unique()):
  winners_team[i]=winning_teams[winning_teams.season==i]['winner'].tail(1).values[0]
winners_of_IPL=pd.Series(winners_team)  
winners_of_IPL=pd.DataFrame(winners_of_IPL,columns=['team'])
```

```python
winners_of_IPL['team'].value_counts().sort_values(ascending=True).tail(10).plot(kind='barh',figsize=(13,8),fontsize=15,color='black')
plt.title("Team that was winner for the most season",fontsize=18,fontweight='bold')
plt.ylabel("Team Name",size=25)
plt.xlabel("No,OF Times the season was won by that team",size=25)
```

***Observation***
Mumbai Indians won the maximum number of seasons
